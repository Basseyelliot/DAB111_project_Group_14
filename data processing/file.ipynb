```python
import pandas as pd 
import sqlite3
```

```python
data = pd.read_excel("../bikes_orderlines.xlsx")
data.head().T
```

```python
data.describe()
```

```python
database_data = data[['order_date', 'order_id', 'quantity', 'price', 'total_price', 'city', 'state']]
```

```python
database_data.head().T
```

```python
database_data.columns
```

```python
conn = sqlite3.connect("../database/bikes.db") 
cur = conn.cursor()

database_data.to_sql("bikes_orderlines", conn, index=False)
conn.commit()
```

```python
rows = cur.execute("SELECT * FROM bikes_orderlines").fetchall()
rows[0]
```
